#**OutLiers (Valores fuera de lo comun, identificarlos y sacarlos con quartiles)**Untitled.png

```python
import  pandas as pandas #Importamos la libreria pandas q es bara ciencia de datos
%matplotlib inline
import matplotlib.pyplot as pyplot

pandas.set_option('display.max_rows', 100) #le digo la cantidad de filas q quiero ver, SI NO TIENE MAS DE 1000 PUES MOSTRARA LAS  Q HALLA
pandas.set_option('display.max_columns', 100) #le digo la cantidad de columnas q quiero ver,SI NO TIENE MAS DE 1000 PUES MOSTRARA LAS  Q HALLA

#CASAS LO GENERO LA FUNCION PASADA
Dataset = pandas.read_csv('SeleccionCasas.csv', sep=';')#aqui le digo q hare un dataset, que leere un archivo .csv y le digo el nombre del archivo (si esta en colab debe volver a  subirse)
#con sep=';' le decimos que las columnas estan SEParadas por un;
Dataset.boxplot(['Valor'], by =['Tipo'])#Me mostrara los Outliers o datos fuera de lo normal , por cada tipo de local(Grafica)
pyplot.show()
#Vemos q hay un dato fuera de lo normal, de 800000

print('Valores iguales o mayores a 800000 ')
print(Dataset[Dataset['Valor']>= 800000])#Asi q veremos cuales son los valores q son iguales o mayores a 800000
print('----------------------------------------------------------------------------------------')
valor= Dataset['Valor']
Q1 =valor.quantile(.25)# q representa el 25%
Q3 =valor.quantile(.75)# q representa el 75%
IIQ = Q3-Q1 #IIQ es inter quartil
limite_inferior = Q1-1.5 * IIQ
limite_superior= Q3+1.5 * IIQ

Seleccion_Valores_Buenos = (valor>=limite_inferior) & (valor<=limite_superior)#aqui me dara una lista de puros true con su posicion, osea si el valor 3 cumplia, abra u ntrue en la posicion 3
Datos_valores_Buenos = Dataset[Seleccion_Valores_Buenos]#con los true, aqui ya filtro los datos
Datos_valores_Buenos.boxplot(['Valor'], by =['Tipo'])#Me mostrara los Outliers o datos fuera de lo normal, por cada tipo de local (Grafica)
pyplot.show()
Datos_valores_Buenos.hist(['Valor'])#en el histograma, veremos q la mayoria de  casas estan en el rango de 1000dolares(Grafica)
pyplot.show()
```

```python
import  pandas as pandas #Importamos la libreria pandas q es bara ciencia de datos
%matplotlib inline
import matplotlib.pyplot as pyplot

pandas.set_option('display.max_rows', 100) #le digo la cantidad de filas q quiero ver, SI NO TIENE MAS DE 1000 PUES MOSTRARA LAS  Q HALLA
pandas.set_option('display.max_columns', 100) #le digo la cantidad de columnas q quiero ver,SI NO TIENE MAS DE 1000 PUES MOSTRARA LAS  Q HALLA

#CASAS LO GENERO LA FUNCION PASADA
Dataset = pandas.read_csv('SeleccionCasas.csv', sep=';')#aqui le digo q hare un dataset, que leere un archivo .csv y le digo el nombre del archivo (si esta en colab debe volver a  subirse)
#con sep=';' le decimos que las columnas estan SEParadas por un;

Area_Graficos = pyplot.figure()#Creara un area para crear varios graficos
Area_Graficos.subplots_adjust(wspace=0.5, hspace=0.5)

#Creare 4 graficos, primero los posicionare en el area
Grafico1 = Area_Graficos.add_subplot(2,2,1)#Agrego el grafico al area ] Linea, columnas, posicion(esta en la 1ra posicion)
Grafico2= Area_Graficos.add_subplot(2,2,2)
Grafico3= Area_Graficos.add_subplot(2,2,3)
Grafico4= Area_Graficos.add_subplot(2,2,4)

#GRAFICO DE DISPERCION
Grafico1.scatter(Dataset.Valor, Dataset.Area)
Grafico1.set_title('Valor y Area')
Grafico1.set_xlabel('Valor')
Grafico1.set_ylabel('Area')

#GRAFICO DE HISTOGRAMA
Grafico2.hist(Dataset.Valor) #NOTA: SE VERA UNNA LINEA POR Q FALTA FILTRAR LOS DATOS, POR Q HAY UNO Q SOBRE PASA EL LLIMITE, OSEA FUERA DEL PROMEDI#
Grafico2.set_title('Histograma')
Grafico2.set_xlabel('Valor')
Grafico2.set_ylabel('Frecuencia')

#GRAFICO DE LINEAS
Datos_Grafico3 = Dataset.Valor.sample(100)#me cogera 100 valores aleatorios de la columna valor
# Resetea el índice pa ordenarlo
Datos_Grafico3.reset_index(drop=True, inplace=True)
Grafico3.plot(Datos_Grafico3)
Grafico3.set_xlabel('Índice')
Grafico3.set_ylabel('Valor')
Grafico3.set_title('Muestra Aleatoria de 100 datos de valor')

#GRAFICO DE BARRAS
Grupo = Dataset.groupby('Tipo')['Valor']#q me los agrupee por tipo, por cada tipo me dara el valor
# en label = Grupo.mean().index, se calcula la media de los valores de 'Valor' para cada grupo único de 'Tipo'. Grupo.mean() calcula estas medias.
#Luego, Grupo.mean().index obtiene las etiquetas de cada grupo de 'Tipo' (que se usarán en el eje X del gráfico)
label = Grupo.mean().index
# Grupo.mean().values obtiene los valores de la media correspondientes a cada grupo (que se usarán para las alturas de las barras en el gráfico).
valores = Grupo.mean().values
Grafico4.bar(label, valores)#crea un gráfico de barras donde label (los diferentes tipos) se usa para las etiquetas del eje X, y valores (las medias de 'Valor') se usan para las alturas de las barras.
Grafico4.set_title('Valor medio por tipo')
Grafico4.set_ylabel('Valor')
# Rotar las etiquetas del eje X para mejorar la legibilidad
Grafico4.set_xticklabels(label, rotation=45, ha="right")
```
